```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/weather_data.csv')
df
```

```python
df['Comfort'] = df['Temperature'] * 0.4 + df['Humidity'] * 0.4 + df['Wind Speed'] * 0.2
```

```python
df.sort_values('Comfort',ascending = False)
```

```python
print(df.loc[df['Comfort'].idxmax(),'Day'],"is the best day to go out based on comfort score")
```

**On analyzing the data using the given formula, Sunday is the best day to go out based on comfort.**

On changing the weightages, if we give more priority to Temperature than Humidity and Wind Speed :-

```python
df['Comfort_New'] = df['Temperature'] * 0.5 + df['Humidity'] * 0.2 + df['Wind Speed'] * 0.3
```

```python
df.sort_values('Comfort_New',ascending = False)
```

```python
print(df.loc[df['Comfort_New'].idxmax(),'Day'],"is the best day to go out based on the new comfort score")
```
