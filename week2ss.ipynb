```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Week2MiniProjectDS.csv")
```

```python
df=pd.DataFrame(data)
df
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)  
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True) 
```

```python
df
```

```python
survived=df["Survived"].value_counts()
survived.plot(kind='pie',labels=['died','survived'],autopct='%1.1f%%')
```

```python
males_survived = df[(df["Sex"] == "male") & (df["Survived"] == 1)].shape[0]
males_survived
males_died=df[(df["Sex"] == "male") & (df["Survived"] == 0)].shape[0]
sizes = [males_survived, males_died]
labels = ['Survived', 'Died']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Survival of Males')
plt.axis('equal') 
plt.show()
```

```python
females_survived = df[(df["Sex"] == "female") & (df["Survived"] == 1)].shape[0]
females_survived
females_died=df[(df["Sex"] == "female") & (df["Survived"] == 0)].shape[0]
sizes = [females_survived, females_died]
labels = ['Survived', 'Died']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Survival of Females')
plt.axis('equal') 
plt.show()
```

```python

sizes = [females_survived, males_survived]
labels = ['females survived', 'males survived']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.axis('equal') 
plt.show()
```

```python
port_survival=df.groupby("Embarked")["Survived"].value_counts().unstack()
port_survival.plot(kind='bar',figsize=(10, 6))
```

we can observe from the above graph that the rate of survival was the highest for pasengers who embarked from Cherbourg.


```python
mean_fare = df.groupby('Survived')['Fare'].mean()
print(mean_fare)
mean_fare.plot(kind='bar')
```

Mean fare was higher for the people who survived.  Therefore we can say that passengers with higher fares were more likely to survive.

```python
def age_group(age):
    if age <= 17:
        return 'Child'
    elif age <= 64:
        return 'Adult'
    else:
        return 'Elderly'
df['AgeGroup'] = df['Age'].apply(age_group)
```

```python
df
```

```python
df.groupby('AgeGroup')["Survived"].value_counts()
```

```python
for age_group in df['AgeGroup'].unique():
    survived_count = df[(df['AgeGroup'] == age_group) & (df['Survived'] == 1)].shape[0]
    total_count = df[df['AgeGroup'] == age_group].shape[0]
    survival_rate = survived_count / total_count if total_count > 0 else 0
    print(f"survival rate for {age_group} ={survival_rate:.2f}")
```

Therefore highest survival rate was of the children onboard.
